```python
import pandas as pd

data = pd.read_csv("data_origin/Culture Offensive Words - Sheet5.csv")
data
```

```python
subset_data = data.loc[data["culture"] == "Chinese"]
subset = []
for i in range(len(subset_data)):
    dataline = subset_data.loc[i]
    data_culture = dataline[0]
    data_category = dataline[1]
    for word in dataline.iloc[2:]:
        subset.append([data_culture, data_category, word])
```

```python
data_clean = pd.DataFrame(subset, columns=["culture", "word_category", "word"])
data_clean["Annotation"] = ""
data_clean
```

```python
data_clean.to_csv("data_clean/Culture Offensive Word - Chinese.csv", index=False)
```
